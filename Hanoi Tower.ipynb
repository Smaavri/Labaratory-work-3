```python
print("Пожалуйста введите колличество дисков для головоломки Ханойская башня:" )
n = int(input())
def hanoi(n, start, finish):
    if n <= 0: return
    temp = 6 - start - finish
    hanoi(n-1, start, temp)
    print("Переносим диск", n, "со", "стержня", start, "на стержень", finish)
    hanoi(n-1, temp, finish)
hanoi(n, 1, 3)
```
